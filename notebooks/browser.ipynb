```python
import gzip
import json
import tensorflow as tf
import collections
import os
from IPython.core.display import display, HTML

tf.logging.set_verbosity(tf.logging.INFO)
```

```python
#PRED_FILE_PATH="bert_model_output/predictions-00-table.json"
#GOLD_PATH = "data/v1.0/dev/nq-dev-00.jsonl.gz"
PREDICTION_PATH = "bert_model_output/predictions-table.json"
GOLD_PATH = "data/filtered_tiny-dev/nq-dev-sample-table.jsonl.gz"
```

```python
from natural_questions.eval_utils import *
#pred_dict = read_prediction_json(PRED_FILE_PATH)
#gold_annotation_dict = read_annotation(GOLD_PATH)

def _open(path):
    if path.endswith(".gz"):
        return gzip.GzipFile(fileobj=tf.gfile.Open(path, "r"))
    else:
        return tf.gfile.Open(path, "r")


def get_html(document_tokens, span):
    
    html = []
    for i in range(span[0], span[1]):
        t = document_tokens[i]
        token = t["token"]
        html.append(token)
        if not t["html_token"]:
            html.append(" ")
    if len(html)>0:
        return "".join(html)
    else:
        return "No Answer"

def print_example_ground_truth(example_t):
    for annot in example_t["annotations"]:
        print("Annotation_id : ", annot["annotation_id"])
        long_answer_span = annot["long_answer"]["start_token"], annot["long_answer"]["end_token"]
        #print("long_answer_span", long_answer_span)
        print("GT Long Answer : ")
        display(HTML(get_html(example_t["document_tokens"], long_answer_span)))
        for answer in annot["short_answers"]:
            short_answer_span = answer["start_token"], answer["end_token"]
            #print("short_answer_span", short_answer_span)
            print("GT Short Answer : ")
            display(HTML(get_html(example_t["document_tokens"], short_answer_span)))
        print("********************")

def print_single_example(e):
    example_id = e["example_id"]
    print("Example id : %d"% example_id)
    print("Title :"+ e["document_title"])
    print("URL : "+ e["document_url"])
    print("Question : " + e["question_text"])
    #Print annotations here
    pred = pred_dict[example_id]
    pred_long_answer_span = (pred["long_answer"]["start_token"], pred["long_answer"]["end_token"])
    print("Predicted Long Answer:")
    pred_short_answer_span = (pred["short_answers"][0]["start_token"], pred["long_answer"]["end_token"])
    display(HTML(get_html(e["document_tokens"], pred_long_answer_span)))
    print("Predicted Short Answer:")
    display(HTML(get_html(e["document_tokens"], pred_short_answer_span)))
    print("-----------------------------------------------------------")
```

```python
# Explore all examples
with open(PREDICTION_PATH) as f:
    pred_json = json.load(f, object_pairs_hook=collections.OrderedDict)["predictions"]
    
pred_dict = {}
for pred in pred_json:
    pred_dict[pred["example_id"]] = pred

gold_doct = {}
with _open(GOLD_PATH) as f:
    for line in f:
        e = json.loads(line, object_pairs_hook=collections.OrderedDict)
        example_id = e["example_id"]
        gold_dict[example_id] = e
        print_single_example(e)
```
